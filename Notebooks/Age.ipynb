```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('Data/ADP/FantasyPros_2015_DST_ADP_Rankings.csv')
df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper','ESPN', 'Yahoo', 'FFC'], axis=1, inplace=True)
df.to_csv('Data/ADP/FantasyPros_2015_DST_ADP_Rankings.csv')
```

```python
for i in range(2015, 2016):
    adr = "Data/ADP/FantasyPros_" + str(i) + "_" 
    end = "_ADP_Rankings.csv"
    te = adr + "TE" + end
    df = pd.read_csv(te)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(te)
    overall = adr + "Overall" + end
    df = pd.read_csv(overall)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(overall)
    k = adr + "K" + end
    df = pd.read_csv(k)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(k)
    qb = adr + "QB" + end
    df = pd.read_csv(qb)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(qb)
    rb = adr + "RB" + end
    df = pd.read_csv(rb)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(rb)
    dst = adr + "DST" + end
    df = pd.read_csv(dst)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(dst)
    wr = adr + "WR" + end
    df = pd.read_csv(wr)
    df.drop(['Team', 'Bye', 'MFL', 'Fantrax', 'RTSports', 'Sleeper', 'FFC'], axis=1, inplace=True)
    df.to_csv(wr)
```

```python
years = []
for i in range(2016, 2022):
    adr = "Data/FantasyRankings/" + str(i) + ".csv"
    df = pd.read_csv(adr)
    #df.fillna(0, inplace=True)
    #df.to_csv(adr)
    years.append(df)
```

```python
import matplotlib.pyplot as plt
for year in years:
    year = year[year['FantPos'] == "RB"]
    year = year[year['Rk'] < 25]
    plt.scatter(year['TD.1'], year[year['FantPos'] == "RB"]['FantPt'])

plt.show()

#First Intersting Showing
#TOUCHDOWNS ARE IMPORTANT
```

```python
for year in years:
    year = year[year['FantPos'] == "WR"]
    year = year[year['Rk'] < 10]
    plt.scatter(year['Yds.2'], year[year['FantPos'] == "RB"]['FantPt'])
plt.show()
#In PPR, you don't have to target pass-catchers
```
